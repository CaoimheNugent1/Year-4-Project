```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
G = nx.read_edgelist("ChicagoRegional.txt", nodetype=int)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Is directed: {G.is_directed()}")
```

### Degree Centrality
Degree centrality measures how many direct connections a node has.
Nodes with high degree centrality are often hubs that connect to many others.

```python
degree_centrality = nx.degree_centrality(G)
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 10 nodes by Degree Centrality:")
print(sorted_degree[:10])
sorted_degree_asc =sorted(degree_centrality.items(), key=lambda x: x[1], reverse=False)
print("Bottom 10 nodes by Degree Centrality:")
print(sorted_degree_asc[:10])
```

### Eigenvector Centrality
Eigenvector centrality measures influence — not just how many connections a node has,
but how connected its neighbors are. It captures “importance” in a network recursively.

```python
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=500, tol=1e-06)
sorted_eigen = sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 10 nodes by Eigenvector Centrality:")
print(sorted_eigen[:10])
sorted_eigen_asc = sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=False)
print("Bottom 10 nodes by Eigenvector Centrality:")
print(sorted_eigen_asc[:10])
```

### PageRank
PageRank (used by Google) measures a node’s importance based on how many links it receives and from how important those linking nodes are.

```python
pagerank = nx.pagerank(G, alpha=0.85)
sorted_pagerank = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)
print("Top 10 nodes by PageRank:")
print(sorted_pagerank[:10])
sorted_pagerank_asc = sorted(pagerank.items(), key=lambda x: x[1], reverse=False)
print("Bottom 10 nodes by PageRank:")
print(sorted_pagerank_asc[:10])
```

### Katz Centrality
Katz centrality is similar to eigenvector centrality but accounts for all paths in the network, diminishing with path length.


```python
# Beta controls the attenuation factor; adjust if convergence issues occur
katz_centrality = nx.katz_centrality(G, alpha=0.005, beta=1.0)
sorted_katz = sorted(katz_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 10 nodes by Katz Centrality:")
print(sorted_katz[:10])
sorted_katz_asc = sorted(katz_centrality.items(), key=lambda x: x[1], reverse=False)
print("Bottom 10 nodes by Katz Centrality:")
print(sorted_katz_asc[:10])
```

```python
degree_sorted = np.sort(list(degree_centrality.values()))[::-1]
eigen_sorted = np.sort(list(eigenvector_centrality.values()))[::-1]
pagerank_sorted = np.sort(list(pagerank.values()))[::-1]
katz_sorted = np.sort(list(katz_centrality.values()))[::-1]

# Create a rank axis (1 = most central node)
ranks = np.arange(1, len(degree_sorted) + 1)

plt.figure(figsize=(10,6))
plt.plot(ranks, degree_sorted, label="Degree Centrality", linewidth=2)
plt.plot(ranks, eigen_sorted, label="Eigenvector Centrality", linewidth=2)
plt.plot(ranks, pagerank_sorted, label="PageRank", linewidth=2)
plt.plot(ranks, katz_sorted, label="Katz Centrality", linewidth=2)

plt.yscale('log')
plt.xlabel("Node Rank (1 = most central)")
plt.ylabel("Centrality Value (log scale)")
plt.title("Comparison of Centrality Measures by Node Rank")
plt.legend()
plt.tight_layout()
plt.show()
```
