```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
# PageRank
pagerank_scores = nx.pagerank(G, alpha=0.85)
sorted_pagerank = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)
print(sorted_pagerank)
```

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)
sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1])
print(sorted_degree_centrality)
```

```python
# Eigenvector Centrality
eigenvector = nx.eigenvector_centrality(G)
sorted_eigenvector = sorted(eigenvector.items(), key=lambda x: x[1])
print(sorted_eigenvector)
```

```python
#Katz Centrality
katz_scores = nx.katz_centrality(G, alpha=0.1, beta=1.0)
sorted_katz_centrality = sorted(katz_scores.items(), key=lambda x: x[1])
print(sorted_katz_centrality)
```

```python
nodes = sorted(G.nodes())
pagerank_values = [pagerank_scores[n] for n in nodes]
degree_values = [degree_centrality[n] for n in nodes]
katz_values = [katz_scores[n] for n in nodes]
eigenvector_values= [eigenvector[n] for n in nodes]

plt.figure(figsize=(12, 6))
plt.plot(nodes, pagerank_values, marker='o', label='PageRank', color='blue')
plt.plot(nodes, degree_values, marker='s', label='Degree Centrality', color='orange')
plt.plot(nodes, katz_values, marker='^', label='Katz Centrality', color='green')
plt.plot(nodes, eigenvector_values, marker='D', label='Eigenvector Score', color='red')

plt.title("Comparison of Centrality Measures â€” Karate Club Graph")
plt.xlabel("Node ID")
plt.ylabel("Centrality Score")
plt.legend()
plt.grid(alpha=0.3)
plt.xticks(nodes)
plt.tight_layout()
plt.show()
```

```python
max_pagerank = max(pagerank_scores.values())

def normalize_to_pagerank(scores, max_ref=max_pagerank):
    max_score = max(scores.values())
    factor = max_ref / max_score
    return {n: score * factor for n, score in scores.items()}

degree_norm = normalize_to_pagerank(degree_centrality)
katz_norm = normalize_to_pagerank(katz_scores)
eigenvector_norm = normalize_to_pagerank(eigenvector)
nodes = sorted(G.nodes())
pagerank_values = [pagerank_scores[n] for n in nodes]
degree_norm_values = [degree_norm[n] for n in nodes]
katz_norm_values = [katz_norm[n] for n in nodes]
eigenvector_norm_values = [eigenvector_norm[n] for n in nodes]

# --- 5. Plot all centralities on same line chart ---
plt.figure(figsize=(12, 6))
plt.plot(nodes, pagerank_values, marker='o', label='PageRank', color='blue')
plt.plot(nodes, degree_norm_values, marker='s', label='Degree Centrality (norm)', color='orange')
plt.plot(nodes, katz_norm_values, marker='^', label='Katz Centrality (norm)', color='green')
plt.plot(nodes, eigenvector_norm_values, marker='d', label='Eigenvector Centrality (norm)', color='red')

plt.title("Comparison of Centrality Measures â€” Normalized to PageRank")
plt.xlabel("Node ID")
plt.ylabel("Centrality Score (Normalized)")
plt.legend()
plt.grid(alpha=0.3)
plt.xticks(nodes)
plt.tight_layout()
plt.show()
```
